# **Chapter 1: The Machine Learning Landscape**

# **Apa Itu Machine Learning?**
Definisi:
Machine Learning (ML) adalah cabang ilmu komputer yang memungkinkan komputer belajar dari data tanpa pemrograman eksplisit.

Contoh:
Filter spam mempelajari pola dari email sebelumnya untuk menentukan apakah email baru termasuk spam atau tidak.

Pendekatan ML:

Tugas (T): Tujuan dari model.

Pengalaman (E): Data yang digunakan.

Performa (P): Metode pengukuran hasil.

# **Mengapa Menggunakan Machine Learning?**
ML sangat berguna dalam:

* Masalah kompleks dengan banyak aturan.
* Tugas yang sulit diprogram secara eksplisit (misalnya, pengenalan suara).
* Lingkungan yang berubah cepat.
* Analisis pola dari data besar (big data).




# **Contoh Aplikasi Machine Learning**
* Klasifikasi gambar: CNN.
* Deteksi objek/tumor: Semantic Segmentation.
* NLP (teks): RNN, Transformer.
* Prediksi nilai: Linear/Polynomial Regression.
* Deteksi anomali: Isolation Forest, One-Class SVM.
* Sistem rekomendasi: Neural Network.
* Game AI: Reinforcement Learning.

# **Jenis Sistem ML**
1. Supervised Learning: Data dilabeli (klasifikasi, regresi).
2. Unsupervised Learning: Data tidak memiliki label (clustering, deteksi anomali).
3. Semi-supervised: Campuran data berlabel dan tidak.
4. Reinforcement Learning: Agen belajar dari reward dan penalti.

Kategori Lainnya:
* Batch vs Online: Belajar sekaligus vs terus-menerus.
* Instance-based vs Model-based: Menghafal vs membangun model.

# **Contoh Kode Regresi Linear**

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Data dummy: X = pendapatan, y = tingkat kepuasan
X = np.array([[10000], [25000], [40000], [60000], [80000]])
y = np.array([5.1, 5.5, 6.0, 7.0, 7.5])

# Membuat model regresi linear
model = LinearRegression()
model.fit(X, y)

# Prediksi kepuasan hidup untuk pendapatan baru
X_pred = [[30000]]
hasil = model.predict(X_pred)

print("Perkiraan kepuasan hidup:", hasil[0])
```

# **Tantangan Umum**
Data:
* Terlalu sedikit
* Tidak representatif
* Berkualitas rendah
* Fitur tidak relevan

Model:
* Overfitting: Model terlalu kompleks.
* Underfitting: Model terlalu sederhana.

Solusi:
* Regularisasi
* Tambah data
* Feature engineering



# **Evaluasi dan Validasi**
* Train/Test Split
* Validation Set
* Train-dev Set
* Cross-validation: Evaluasi dengan pembagian data berkali-kali.



# **No Free Lunch Theorem**
Tidak ada algoritma terbaik untuk semua masalah. Pilihan model bergantung pada asumsi dan jenis data.
